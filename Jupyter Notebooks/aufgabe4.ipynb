```python
"""
In this section, we define the starting conditions for simulating the oscillation of the mechanical system. 
These values represent the initial state of the system before it begins to move. By setting these parameters, 
we ensure that the simulation starts from a well-defined position and behaves as expected.
"""

START_DEFLECTION = math.pi / 18  # initial angular displacement
START_VELOCITY = 0  # initial angular velocity
MASS = 1  # mass of the oscillating system 
DEFAULT_C = 5.0  # default spring constant
DEFAULT_D = 0.5  # default damping coefficient
```

```python
"""
In this cell, we will create a class for the IntSolver. We will then create an `IntSolver` instance, which will
numerically solve the system's motion differential equation using the `odeint` function from the `scipy.integrate`
library.

The class has two methods:
1. `calculate`: Defines the differential equation of the system.
2. `integrate`: Solves the differential equations over time using the initial conditions and system parameters.

Example Usage:
--------------
# Create an instance of the IntSolver class
solver = IntSolver()

# Define the time array for simulation
import numpy as np
t = np.linspace(0, 10, 1000)  # Time from 0 to 10 seconds, divided into 1000 steps

# Define system parameters
c = 5.0  # spring constant
d = 0.5  # damping coefficient
m = 1.0  # mass

# Solve the system's motion
solution = solver.integrate(func=solver.calculate, start_deflection=math.pi/18, start_velocity=0, t=t, c=c, d=d, m=m)
"""

class IntSolver:
    def __init__(self) -> None:
        return None

    def calculate(self, x, t, c, d, m):
        """Calculates the derivatives of the system's state variables.

        Args:
            x: A tuple containing the current state variables of the system.
                        - x[0]: Angular displacement (phi).
                        - x[1]: Angular velocity (phi_dot).
            t: The current time. This is not used in the equation but is required by the solver.
            c: The spring constant (stiffness of the spring).
            d: The damping coefficient (resistance to motion due to damping).
            m: The mass of the system.

        Returns:
            tuple: A tuple containing the derivatives of the state variables.
                   - x_dot[0]: The derivative of the angular displacement (angular velocity).
                   - x_dot[1]: The derivative of the angular velocity (angular acceleration). 
        """
        x_1, x_2 = x[0], x[1]  # unpack the state variables: x_1 = phi, x_2 = phi_dot

        # Write your code for the differential equation here ⬇️
        # x_dot = ... 
        x_dot = (x_2, -d / m * x_2 - c / m * x_1)  # x_dot = (x_2, -d / m * x_2 - c / m) 
        return x_dot

    def integrate(self, func, start_deflection, start_velocity, t, *args):
        """This method solves the system's motion by numerically integrating the differential equations
            defined in the `calculate` method. It uses the initial conditions and system parameters to
            compute the system's angular displacement and velocity over time.

        Args:
            func: The function that defines the differential equations (in this case, the `calculate` method).
            start_deflection: The initial angular displacement of the system.
            start_velocity: The initial angular velocity of the system.
            t: A time array over which the integration is performed.
            *args: Additional arguments to pass to the `func` (e.g., spring constant `c`, damping coefficient `d`,
                mass `m`).

        Returns:
            array: The solution to the differential equations, containing the angular displacement
                and velocity over time. Each row corresponds to a time step, and the columns represent the state
                variables (phi, phi_dot)
        """
        y_0 = (start_deflection, start_velocity)
        x = odeint(func=func, y0=y_0, t=t, args=args)

        return x


# make an instance of IntSolver
calculator = IntSolver()
```
