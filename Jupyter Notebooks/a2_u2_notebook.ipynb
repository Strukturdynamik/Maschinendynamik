A bar (moment of inertia $J$ with respect to $A$) is pivoted at point $A$. At the end of the bar, a linear spring-damper element (spring stiffness $c$, damping constant $d$) is attached, which is subjected to a periodic motion $u(t) = \hat{u} \cos(\Omega t)$ at its base.

### Tasks
1. Isolate the body in its displaced position and indicate all forces and moments acting on it.

2. Derive the equation of motion for the system for small oscillations around the equilibrium position $\varphi = 0$. Use the general form:

   $$
   \ddot{\varphi} + 2\delta \dot{\varphi} + \omega_{0}^{2}\varphi 
   = b_{0}u + b_{1}\dot{u} + b_{2}\ddot{u}
   $$

3. Determine the amplitude response $V(\Omega)$ and sketch it.

**Given:** $J_{a}, c, d, l, u(t)$

```python
import numpy as np
J_A = 100
L = 10
C = 50
D = 0.5
U_HAT = 0.05
PHI_0 = np.pi * 0 / 180
PHI_0_DOT = 0
OMEGA_0 = np.sqrt(
    (C * L**2) / J_A
)
OMEGA = OMEGA_0
ALPHA = OMEGA / 10
```

```python
import numpy as np
from scipy.integrate import odeint

from src.project.calculations.validate_solution import validate_solution

correct = False

class StudentSolver:
    def __init__(self) -> None: 
        return None 

    # Dauerlauf
    def state_space_steady(self, x, t, l, j_a, u_hat, d, c, omega):
        delta = (d * l**2) / (2 * j_a)
        b1 = (d * l) / j_a
        b0 = (c * l) / j_a
        omega_0 = np.sqrt((c * l**2) / j_a)

        [phi, phi_dot] = x
        
        # Write your code here ⬇️
        x_p = None
        return x_p

    # Hochlauf
    def state_space_accelerated(self, x, t, l, j_a, u_hat, d, c, alpha):
        delta = (d * l**2) / (2 * j_a)
        b1 = (d * l) / j_a
        b0 = (c * l) / j_a
        omega_0 = np.sqrt((c * l**2) / j_a)

        [phi, phi_dot] = x
        x_p = [
            phi_dot,
            -2 * delta * phi_dot
            - omega_0**2 * phi
            + b0 * u_hat * np.cos(0.5 * alpha * t**2)
            - b1 * u_hat * alpha * t * np.sin(0.5 * alpha * t**2),
        ]
        return x_p

    def integrate(self, func, phi_0, phi_0_dot, t, *args):
        x0 = (phi_0, phi_0_dot)
        return odeint(func=func, y0=x0, t=t, args=args)[:, 0]
    


from src.project.calculations.int_solver_a2_u2 import IntSolverAufgabe2Uebung2 as IntSolverCorrect

np.set_printoptions(precision=4, suppress=True, linewidth=120)
student_solver = StudentSolver()
correct_solver = IntSolverCorrect()
t = np.linspace(0, 20, 500)


# settled
settled_student = student_solver.integrate(
                    student_solver.state_space_steady,
                    PHI_0,
                    PHI_0_DOT,
                    t,
                    L,
                    J_A,
                    U_HAT,
                    D,
                    C,
                    OMEGA,
                )

settled_correct = correct_solver.integrate(
                    student_solver.state_space_steady,
                    PHI_0,
                    PHI_0_DOT,
                    t,
                    L,
                    J_A,
                    U_HAT,
                    D,
                    C,
                    OMEGA,
                )

# accelerated
accelerated_student = student_solver.integrate(
                    student_solver.state_space_accelerated,
                    PHI_0,
                    PHI_0_DOT,
                    t,
                    L,
                    J_A,
                    U_HAT,
                    D,
                    C,
                    ALPHA,
                )
accelerated_correct = correct_solver.integrate(
                    student_solver.state_space_accelerated,
                    PHI_0,
                    PHI_0_DOT,
                    t,
                    L,
                    J_A,
                    U_HAT,
                    D,
                    C,
                    ALPHA,
                )


# test results
print("Settled: ")
correct_settled = validate_solution(settled_correct, settled_student, relative_threshold=0.05)
print()
print("Accelerated: ")
correct_acc = validate_solution(accelerated_correct, accelerated_student, relative_threshold=0.05)

# solution is correct if both settled and accelerated are correct
correct = correct_settled and correct_acc
```
