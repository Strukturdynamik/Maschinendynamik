A massless rod is pivoted at point $A$ (rotation angle $\varphi$). Along the rod, a block (mass $m$, moment of inertia $J_{s}$ with respect to $S$) can slide. The block is attached to a spring (spring constant $c$, unstretched spring length $s = s_{0}$). Frictional effects are neglected. The minimal coordinates are chosen as $q_{1} = s$ and $q_{2} = \varphi$.

### Tasks
Determine the
1. kinetic energy $T(\mathbf{q})$.

2. potential energy $U(\mathbf{q})$.

3. nonlinear equations of motion using the Lagrange equations of the 2nd kind.

4. equilibrium positions.



In the following, the behavior of the system for small displacements around the equilibrium $\varphi_{R} = 0$ and the corresponding value $s_{R}$ is to be investigated:

5. Derive the linearized equations of motion in the form $\mathbf{M\ddot{q} + K q = 0}$.
   

6. Determine the mass matrix $\mathbf{M}$ from the kinetic energy.

7. Determine the stiffness matrix $\mathbf{K}$ from the potential energy.



**Given:** $J_{s}, m, c$
