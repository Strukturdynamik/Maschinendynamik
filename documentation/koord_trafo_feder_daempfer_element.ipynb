# Standbild

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 10, 1001)
phi = 20*np.cos(5*t)
phi = np.deg2rad(phi)

#plt.figure(figsize=(14,8))
#plt.subplot(1,2,1)
#plt.plot(t, phi);


#plt.subplot(1,2,2)
plt.axis([-2, 4, -3, 3])

# Stange
plt.plot([0, 2.5*np.cos(phi[0])], [0, 2.5*np.sin(phi[0])], color='black');

# Transformationswinkel
alpha = np.arctan2(np.cos(phi)-1, np.sin(phi)-1.3)

# obere Transformationmatrix
K_o = np.array([[np.sin(alpha[0]), -np.cos(alpha[0]), 1-0.355*np.sin(alpha[0])],
                [np.cos(alpha[0]),  np.sin(alpha[0]), 1.3-0.355*np.cos(alpha[0])],
                [0, 0, 1]])

# Ortsvektor des KOS-Ursprungs 2 in globalen Koordinaten
r_O2_1 = np.array([1-0.355*np.sin(alpha[0]), 1.3-0.355*np.cos(alpha[0]), 1])

# oberer Gabelungspunkt
r_P1_2 = np.array([-0.35, 0, 1])

# linker Gabelungspunkt
r_P2_2 = np.array([0, 0.2, 1])

# rechter Gabelungspunkt
r_P3_2 = np.array([0, -0.2, 1])

# untere Transformationsmatrix
K_u = np.array([[np.sin(alpha[0]), -np.cos(alpha[0]), np.cos(phi[0])+0.1*np.sin(alpha[0])],
                [np.cos(alpha[0]),  np.sin(alpha[0]), np.sin(phi[0])+0.1*np.cos(alpha[0])],
                [0, 0, 1]])

# Ortvektor des KOS-Ursprungs 3 in globalen Koordinaten
r_O3_1 = np.array([np.cos(phi[0])+0.1*np.sin(alpha[0]), np.sin(phi[0])+0.1*np.cos(alpha[0]), 1])

# unterer Gabelungspunkt
r_P4_3 = np.array([-0.1, 0, 1])

# linker Gabelungspunkt
r_P5_3 = np.array([0, 0.2, 1])

# rechter Gabelungspunkt
r_P6_3 = np.array([0, -0.2, 1])

# Punkte in globalen Koordinaten
r_P1_1 = np.matmul(K_o, r_P1_2)
r_P2_1 = np.matmul(K_o, r_P2_2)
r_P3_1 = np.matmul(K_o, r_P3_2)
r_P4_1 = np.matmul(K_u, r_P4_3)
r_P5_1 = np.matmul(K_u, r_P5_3)
r_P6_1 = np.matmul(K_u, r_P6_3)

plt.plot([r_P4_1[0], r_O3_1[0]], [r_P4_1[1], r_O3_1[1]], color='black')
plt.plot([r_P5_1[0], r_P6_1[0]], [r_P5_1[1], r_P6_1[1]], color='black')
plt.plot([r_O2_1[0], r_P1_1[0]], [r_O2_1[1], r_P1_1[1]], color='black')
plt.plot([r_P2_1[0], r_P3_1[0]], [r_P2_1[1], r_P3_1[1]], color='black')

plt.show()

# Elemente noch nicht rechtwinklig, weil Achsen unterschiedliche Stauchung haben
```
